```python
from lxml import etree
```

```python
# Parse the XML file
book_files = xmls[2]
book_folder = xmls[0]
book_files.sort()

joined_text = ""
for f in book_files:
    #print(f)
    tree = etree.parse(os.path.join(book_folder, f))
    # Get the root element
    root = tree.getroot()
    
    # Print the root tag
       
    # Define the namespace (replace with the correct URL from your XML if different)
    namespace = {'ns': 'http://schema.primaresearch.org/PAGE/gts/pagecontent/2013-07-15'}
    
    # Find all <Unicode> elements within the specified namespace
    unicode_elements = tree.findall('.//ns:Unicode', namespaces=namespace)
    
    # Extract the text content of each <Unicode> element and join them as one long string
    joined_text += '\n'.join([element.text for element in unicode_elements if element.text != None])
```
